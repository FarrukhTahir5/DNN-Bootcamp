```python
# def cubic_exp(A):
#     res = []
#     for num in A:
#         res.append(num+num**(0.5))
#     return res
def cubic_exp(A):
    res = []
    for i in range(len(A)):
        if A[i] <= 8:
            res.append(1 * A[i])  # Straight line with slope 0.5
        else:
            res.append(0.5 * A[i] + (A[i] - 8)**2)  # Becomes more vertical after A > 8
    return res
A=[1,2,3,4,5,6,16,32]
B=[1,2,3,4,5,6,7,8]
C=cubic_exp(A)
plt.scatter(B,A)
plt.plot(A,C,c='r')
plt.xlim(0,30)
plt.ylim(0,30)
plt.show()
```
